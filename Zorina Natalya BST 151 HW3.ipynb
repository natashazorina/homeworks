```python
# Задача 1
import requests
from urllib import parse as urlparse
from bs4 import BeautifulSoup, SoupStrainer
 
 
def print_page_info(host, path):
    page = requests.get(host + path)
    soup = BeautifulSoup(page.text, "html.parser")
    print('Title: "{}", length: {}'.format(soup.title.string, len(soup.get_text())))
 
 
def is_relative(url):
    return not bool(urlparse.urlparse(url).netloc)
 
 
host = "https://ru.wikipedia.org"
path = "/wiki/Всемирная_библиотека_(Норвежский_книжный_клуб)"
page = requests.get(host + path)
soup = BeautifulSoup(page.text, "html.parser", parse_only=SoupStrainer("a"))
 
for tag in soup:
    if tag.has_attr("href"):
        link = tag["href"]
        if is_relative(link):
            print_page_info(host, link)
```

```python
# Задача 2
import requests
from urllib import parse as urlparse
from bs4 import BeautifulSoup, SoupStrainer
 
 
def is_relative(url):
    return not bool(urlparse.urlparse(url).netloc)
 
 
def find_first_link(page):
    soup = BeautifulSoup(page.text, "html.parser", parse_only=SoupStrainer("a"))
    for tag in soup:
        if tag.has_attr("href"):
            link = tag["href"]
            if is_relative(link):
                return "https://ru.wikipedia.org/" + link
 
 
def wiki_walk(url, N):
    if N > 0:
        page = requests.get(url)
        soup = BeautifulSoup(page.text, "html.parser")
        print('Title: "{}"'.format(soup.title.string))
        wiki_walk(find_first_link(page), N - 1)
 
print ('введите ссылку') 
url = input()
print ('введите N')
N = input()
wiki_walk(url, int(N))
```
